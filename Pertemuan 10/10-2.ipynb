```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = [8,5]
plt.rcParams['font.size'] = 14
plt.rcParams['font.weight'] = "bold"
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
path = ",,/input"
df = pd.read_csv('Data_Asuransi.csv')
print('Number of Rows and Columns in the Datasets', df.shape)
print('')
df.head()
```

```python
sns.lmplot(x="bmi", y="charges", data=df, aspect=2, height=6)
plt.xlabel('Body Mass Index $(Kg/m^2)$: as Independent Variable')
plt.ylabel('Insurance Charger: as Dependent Variable')
plt.title("Charge vs BMI");
```

```python
df.describe()
```

```python
plt.figure(figsize=(12,4))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis", yticklabels=False)
plt.title('Missing Values in Datasets');
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, cmap="Wistia", annot=True);
```

```python
f = plt.figure(figsize=(14,6))

ax =f.add_subplot(121)
sns.violinplot(x='sex', y='charges', data=df, palette="Wistia", ax=ax);
ax.set_title('Violin Plot of Charges VS Sex')

ax =f.add_subplot(122)
sns.violinplot(x='smoker', y='charges', data=df, palette="magma", ax=ax);
ax.set_title('Violin Plot of Charges VS Smoker')
```

```python
plt.figure(figsize=(14,6))
sns.boxplot(x='children', y='charges', data=df, hue='sex', palette='rainbow')
ax.set_title('Box Plot of Charges VS Children')
```

```python
plt.figure(figsize=(14,6))
sns.violinplot(x='region', y='charges', data=df, palette="rainbow", split=True)
ax.set_title('Violin Plot of Charges VS Region')
```

```python
df.groupby('children').agg(['mean', 'min', 'max'])['charges']
```

```python
f = plt.figure(figsize=(14,6))

ax =f.add_subplot(121)
sns.scatterplot(x='age', y='charges', data=df, palette="magma", hue='smoker', ax=ax)
ax.set_title('Violin Plot of Charges VS Age');

ax =f.add_subplot(122)
sns.scatterplot(x='bmi', y='charges', data=df, palette="viridis", hue='smoker')
ax.set_title('Violin Plot of Charges VS BMI');
plt.savefig('sc.png')
```

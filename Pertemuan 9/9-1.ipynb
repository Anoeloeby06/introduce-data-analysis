```python
import seaborn as sns
sns.set(style="darkgrid")
ax = sns.countplot(x = "Wilayah", data = df)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (5,5))

ax = sns.countplot(x = "Wilayah", data = df)
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2., height,"{:1.0f}".format(height), ha = "center")
```
